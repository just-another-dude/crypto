```python
############################
### prepare api from CMC ###
############################
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
import pandas as pd

url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
parameters = {
  'start':'1',
  'limit':'5000',
}
headers = {
  'Accepts': 'application/json',
  'X-CMC_PRO_API_KEY': mykey,
}

session = Session()
session.headers.update(headers)
```

```python
#########################
### read api from CMC ###
#########################
response = session.get(url, params=parameters)
data = json.loads(response.text)
```

```python
##############################
### BTC example (for data) ###
##############################
data['data'][0]
```

```python
############################
### show intresting data ###
############################
ccoinname = data['data'][0]['name']
ccoinsymbol = data['data'][0]['symbol']
ccoinvalue = data['data'][0]['quote']['USD']['price']
ccoinvolume24 = data['data'][0]['quote']['USD']['volume_24h']
ccoincirculating_supply = data['data'][0]['circulating_supply']
ccoincmc_rank = data['data'][0]['cmc_rank']
print("ccoinname:",ccoinname)
print("ccoinsymbol:",ccoinsymbol)
print("ccoinvalue:",ccoinvalue)
print("ccoinvolume24:",ccoinvolume24)
print("ccoincirculating_supply:",ccoincirculating_supply)
print("ccoincmc_rank:",ccoincmc_rank)
```

```python
##################################
### get all coins map from api ###
##################################
url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/map'

headers = {
  'Accepts': 'application/json',
  'X-CMC_PRO_API_KEY': mykey,
}

session = Session()
session.headers.update(headers)
```

```python
###############################################
### at the moment work only with 5000 coins ###
###############################################
#response = session.get(url)
#dataID = json.loads(response.text)
#coinarrsize = len(dataID['data'])
coinarrsize = 5000
```

```python
######################################################
### sort all coins with relevent info in dataframe ###
######################################################
import pandas as pd
column_names = ["name","symbol","value","volume24","circulating_supply","cmc_rank"]
df = pd.DataFrame(columns=column_names)
for i in range(0,coinarrsize):
    ccoinname = data['data'][i]['name']
    ccoinsymbol = data['data'][i]['symbol']
    ccoinvalue = data['data'][i]['quote']['USD']['price']
    ccoinvolume24 = data['data'][i]['quote']['USD']['volume_24h']
    ccoincirculating_supply = data['data'][i]['circulating_supply']
    ccoincmc_rank = data['data'][i]['cmc_rank']
    df.loc[i] = [ccoinname,ccoinsymbol,ccoinvalue,ccoinvolume24,ccoincirculating_supply,ccoincmc_rank]
    
```

```python
#####################
### show all data ###
#####################
pd.set_option('display.max_rows', None)
df.sort_values(by=['value'])
#pd.set_option('display.max_rows', 10)
```

```python
#################################
### show all data after query ###
#################################
dfunder1 = df[(df['value'] < 0.1) & (df['volume24'] > 1000000) & (df['cmc_rank'] <180) ]
dfunder1.sort_values(by=['circulating_supply'])
```
